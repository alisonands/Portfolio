# Covid-19 Canada tracker
Data gathered from [covid19tracker](https://api.covid19tracker.ca/docs/1.0/overview)

```python
import requests
import pandas as pd
```

```python
url = 'https://api.covid19tracker.ca/reports'
data_raw = pd.DataFrame(requests.get(url).json())['data']
```

```python
data = pd.DataFrame(data_raw.to_list())
data
```

```python
for col in data.columns:
    print(col, 'INT,')
```

```python
data.to_csv('covid.csv')
```

```python
data['date'] = pd.to_datetime(data['date'])
```
