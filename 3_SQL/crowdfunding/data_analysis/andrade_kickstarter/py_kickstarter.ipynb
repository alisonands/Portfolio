```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = np.genfromtxt('/Users/alisonandrade/Documents/alison18/python/kickstarter.csv', delimiter = ',', skip_header = 1)
days = data[:,0]
pledged = data[:,1]
goal_surpassed = data[:,2]
plt.plot(days, pledged, '.')
```

```python
plt.rcParams.update({'font.size': 22})
plt.figure(figsize = (15, 7))
plt.hist(days, bins = round(len(days)/25), label = 'campaign counts')
plt.xlabel('Length of Campaign(days)')
plt.legend()
```

```python
df = pd.read_csv('/Users/alisonandrade/Documents/alison18/python/campaign_country.csv')
plt.rcParams.update({'font.size': 12})
plt.figure(figsize = (15, 7))

grouped_data = df.groupby(['country_name', 'outcome']).sum().reset_index()

pivot_data = grouped_data.pivot(index='country_name', columns='outcome', values='campaign_count')

pivot_data.plot(kind='bar', stacked=True)

plt.xlabel('Country')
plt.ylabel('Campaign Count')

```

```python
df = pd.read_csv('/Users/alisonandrade/Documents/alison18/python/campaign_currency.csv')
plt.rcParams.update({'font.size': 12})
plt.figure(figsize = (15, 7))

grouped_data = df.groupby(['currency_name', 'outcome']).sum().reset_index()

pivot_data = grouped_data.pivot(index='currency_name', columns='outcome', values='campaign_count')

pivot_data.plot(kind='bar', stacked=True)

plt.xlabel('Currency')
plt.ylabel('Campaign Count')

```
