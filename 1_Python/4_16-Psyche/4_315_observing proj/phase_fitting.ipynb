```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
data1 = np.genfromtxt("table_Phase-Folded-Data.csv", delimiter = ',', usecols = range(0,3), skip_header = 1)

plt.figure(figsize = (15, 7))
xfit = np.linspace(0,2,50)
fit = 0.1*np.sin(6.9*xfit)-0.55

plt.plot(data1[:,1], data1[:,2], '.', label = 'data')
plt.plot(xfit, fit, label = 'fit')
plt.title('Phase Folded Light Curve')
plt.xlabel('Phase')
plt.ylabel('Magnitude')
plt.legend()
```

```python
def fit(x, a, b, c):
    y = a*np.sin(b*x)+c
    return y

x = np.linspace(0, 2, 50)

popt, pcov = curve_fit(fit, data1[:,1], data1[:,2])
fit_a = popt[0]
fit_b = popt[1]
fit_c = popt[2]

cos_fit = fit(data1[:,1], fit_a, fit_b, fit_c)

plt.plot(data1[:,1], data1[:,2], 'o', label='data')
plt.plot(data1[:,1], cos_fit, '-', label='fit')
plt.legend()
popt
```
