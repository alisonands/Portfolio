```python
years = 30
def cagr(beginning_value, ending_value, years):
    cagr_val = (ending_value / beginning_value) ** (1 / years) - 1
    return cagr_val
ticker = yf.Ticker("AAPL")
start_val = ticker.history(period="30y")['Open'].iloc[0]
end_val = ticker.history(period="30y")['Open'].iloc[-1]
print(f'the compound annual growth rate for MSFT is {cagr(start_val, end_val, years)}')

dateindex = pd.to_datetime(ticker.history(period="30y").index)
start_time = dateindex[0]
end_time = dateindex[-1]

ticker.info.keys()
print(f'the EV/EBITDA is {ticker.info["enterpriseToEbitda"]}')
print(f'The calculated EV/EBITDA is {ticker.info["enterpriseValue"]/ticker.info["ebitda"]}')
```
